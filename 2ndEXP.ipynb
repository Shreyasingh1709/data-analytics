```python
import matplotlib.pyplot as plt
import numpy as np
#'#' is not a valid marker style in Matplotlib.
#Valid marker options include 'o', 's', '^', '*', etc.
#plt.scatter(x, y) requires 1D arrays
#x and y are 2D arrays of shape (2,5), but plt.scatter() expects 1D arrays.
#You need to flatten them using .flatten().
x = np.random.randint(1,100,(1,5))
print("X:")
print(x)
y = np.random.randint(1,100,(1,5))
print("Y:")
print(y)
#shape of x and y should be same
plt.scatter(x,y,color = 'brown',marker = '*') 
plt.title("Scatter plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#generating random floating-point numbers is np.random.uniform().
x = np.random.randfloat(1,100,(2,5))
print("X:")
print(x)
#np.random.randn() does not accept a (low, high, size) format.
#If you want random values in a range, use np.random.uniform(low, high, size) instead.
y = np.random.randn(1,100,(2,5))
print("Y:")
print(y)
#shape of x and y should be same
plt.scatter(x,y) 
plt.title("Scatter plot")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```
